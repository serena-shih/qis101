```python
"""complex_lattice.ipynb"""
# Cell 1

# Scatter plot for complex function

from __future__ import annotations
import typing
import matplotlib.pyplot as plt
import numpy as np
if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from numpy.typing import NDArray

def plot(ax: Axes) -> None:
    """Plot scatter plot"""
    x_real: NDArray[np.float_] = np.linspace(-4, 4, 100)
    x_imag: NDArray[np.float_] = np.linspace(0, 2, 100)
    x: NDArray[np.complex_] = []
    for a in x_real:
        for b in x_imag:
            x = np.append(x, a + 1j * b)
    y: NDArray[np.complex_] = x**2 + 1j * x + 1

    ax.scatter(np.real(y), np.imag(y), 1)
    ax.set_title("$f(z) = z^2 + iz + 1$")
    ax.set_xlabel("Real")
    ax.set_ylabel("Imaginary")
    ax.set_aspect("equal")

plot(plt.axes())
```

```python
# Cell 2

# Count number of Gaussian integers with |Re(f(z))| and |Im(f(z))| <= 10

count: int = 0
z: np.complex_ = 0
f: np.complex_ = 0
for x in range(-500, 500):
    for y in range(-500, 500):
        z = x + 1j * y
        f = z**2 + 1j * z + 1
        if (abs(np.real(f)) <= 10) and (abs(np.imag(f)) <= 10):
            count += 1
print(f"There are {count} Gaussian integers.")
```
